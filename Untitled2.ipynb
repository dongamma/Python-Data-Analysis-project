```python
from geopy.geocoders import Nominatim
import pandas as pd

def geocoding(address):
    geocoder = Nominatim(user_agent='South Korea', timeout=None)
    location = geocoder.geocode(address)
    if location is not None:
        return {'위도': location.latitude, '경도': location.longitude}
    else:
        return {'위도': None, '경도': None}


# '단속장소' 열의 주소 값을 좌표로 변환하여 '위도'와 '경도' 열 추가
iwp_select2[['위도', '경도']] = iwp_select2['단속장소'].apply(geocoding).apply(pd.Series)

# 결과 출력
print(iwp_select2)
```
