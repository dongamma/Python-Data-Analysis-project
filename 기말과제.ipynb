- 그룹관리번호가 서로 같다면 같은 주차장임

```python
Road_Park_Data = ['주차장고유값(pl_id)', '법정동코드(bemd_code)', '그룹관리번호(managing_group_no)' , '공유주차장구분(shared_parkinglot)', '경도(longitude)', '위도(latitude)']
Road_Park_Select = Road_Park[Road_Park_Data]
Road_Park_Select
```

코드말고 이름으로 바꿔주기

```python
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10500, '법정동코드(bemd_code)'] = '자양동'
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10100, '법정동코드(bemd_code)'] = '중곡동'
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10300, '법정동코드(bemd_code)'] = '구의동'
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10400, '법정동코드(bemd_code)'] = '광장동'
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10700, '법정동코드(bemd_code)'] = '화양동'
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10200, '법정동코드(bemd_code)'] = '능동'
Road_Park_Select.loc[Road_Park_Select['법정동코드(bemd_code)'] == 10900, '법정동코드(bemd_code)'] = '군자동'
Road_Park_Select
```

그래프로 시각화

```python
park_df2 = Road_Park_Select['법정동코드(bemd_code)'].value_counts()
```

```python
park_df2.plot(kind='bar')
```

그룹 관리 번호 중복을 방지, 실질적으로 350개의 크고작은 주차장이 있다는걸 알 수 있음 

```python
Road_Park_Select_unique = Road_Park_Select.drop_duplicates(subset='그룹관리번호(managing_group_no)')
Road_Park_Select_unique
```

노상주차장 중에도 공유주차장은 단하나

```python
df_filtered = (Road_Park_Select.loc[pd.notnull(Road_Park_Select['공유주차장구분(shared_parkinglot)'])]
               .drop_duplicates(subset='그룹관리번호(managing_group_no)'))
df_filtered
```

광진구에 존재하는 노상주차장을 파이썬folium을 이용해 지도로 시각화

```python
m = folium.Map(location=[Road_Park_Select_unique['위도(latitude)'].mean(), Road_Park_Select_unique['경도(longitude)'].mean()], zoom_start=13)
```

```python
for index, row in Road_Park_Select_unique.iterrows():
    folium.Marker([row['위도(latitude)'], row['경도(longitude)']]).add_to(m)
```

```python
m
```

전체적으로 봤을때 광진구에서 주차장이 가장 많은 3군데는 중곡동, 자양동, 구의동 순이다

```python
park_df1 = BuPark_filtered['법정동코드(bemd_code)'].value_counts().reset_index()
park_df1.columns = ['법정동코드(bemd_code)', '주차장 수']

park_df2 = Road_Park_Select['법정동코드(bemd_code)'].value_counts().reset_index()
park_df2.columns = ['법정동코드(bemd_code)', '주차장 수']

combined_df = pd.merge(park_df1, park_df2, on='법정동코드(bemd_code)')
combined_df['주차장 수'] = combined_df['주차장 수_x'] + combined_df['주차장 수_y']
combined_df.drop(['주차장 수_x', '주차장 수_y'], axis=1, inplace=True)

combined_df
```

# 광진구 인구 현황 파악

```python
people = pd.read_csv('인구현황.csv', encoding='cp949')
people
```

맨위의 행은 필요없으니 삭제하고 생기는 NaN값 정리

```python
people = people.shift(-1).dropna()
people
```

```python
people.info()
```

```python
people.describe().T
```

전체적으로 어디가 인구수가 제일 많은지 분석

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')

x = people['행정기관']
y = people['계']

sorted_index = y.argsort()[::-1]

x_sorted = x[sorted_index]
y_sorted = y[sorted_index]

plt.bar(x_sorted, y_sorted)

plt.title('인구수')
plt.xlabel('행정기관')
plt.ylabel('계')

plt.xticks(rotation=90)
plt.show()
```
